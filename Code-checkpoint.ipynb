## Load the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# 1. Load the Dataset
list_of_orders = pd.read_csv('List_of_Orders.csv')
list_of_orders.head()

import warnings
warnings.filterwarnings('ignore')
```

```python
order_details = pd.read_csv('Order_Details.csv')
order_details.head()
```

```python
sales_target = pd.read_csv('Sales_Target.csv')
sales_target.head()
```

```python
# Inspect the column names
print("List of Orders Columns:", list_of_orders.columns)
print("Order Details Columns:", order_details.columns)
print("Sales Target Columns:", sales_target.columns)
```

##  Data Preprocessing and Cleaning

```python
# Check for missing values
print("Missing values in List of Orders:\n", list_of_orders.isnull().sum())
```

```python
print("Missing values in Order Details:\n", order_details.isnull().sum())
```

```python
# Convert 'Month of Order Date' to datetime format in sales_target
if 'Month of Order Date' in sales_target.columns:
    try:
        sales_target['Month of Order Date'] = pd.to_datetime(sales_target['Month of Order Date'], format='%b-%y')
    except ValueError as e:
        print(f"Error parsing 'Month of Order Date' in sales_target: {e}")
else:
    print("Column 'Month of Order Date' not found in sales_target. Available columns:", sales_target.columns)
```

## Feature Engineering

```python
# Merge datasets to get a complete view
merged_data = pd.merge(list_of_orders, order_details, on='Order ID')
```

```python
# Create RFM (Recency, Frequency, Monetary) features
latest_date = merged_data['Order Date'].max()
merged_data['Recency'] = (latest_date - merged_data['Order Date']).dt.days
rfm = merged_data.groupby('CustomerName').agg({
    'Recency': 'min',
    'Order ID': 'count',
    'Amount': 'sum'
}).reset_index()
rfm.columns = ['CustomerName', 'Recency', 'Frequency', 'Monetary']
```

## Clustering Algorithms for Segmentation

```python
# Standardize the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

## K-Means Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['KMeans_Cluster'] = kmeans.fit_predict(rfm_scaled)
kmeans_silhouette = silhouette_score(rfm_scaled, rfm['KMeans_Cluster'])
print(f'K-Means Silhouette Score: {kmeans_silhouette}')
```

## DBSCAN Clustering

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
rfm['DBSCAN_Cluster'] = dbscan.fit_predict(rfm_scaled)
dbscan_silhouette = silhouette_score(rfm_scaled, rfm['DBSCAN_Cluster'], metric='euclidean') if len(set(dbscan.labels_)) > 1 else -1
print(f'DBSCAN Silhouette Score: {dbscan_silhouette}')
```

## Agglomerative Clustering

```python
agglo = AgglomerativeClustering(n_clusters=4)
rfm['Agglo_Cluster'] = agglo.fit_predict(rfm_scaled)
agglo_silhouette = silhouette_score(rfm_scaled, rfm['Agglo_Cluster'])
print(f'Agglomerative Clustering Silhouette Score: {agglo_silhouette}')
```

## Evaluating and Interpreting Clusters

```python
# Add cluster labels to the original data
merged_data = pd.merge(merged_data, rfm[['CustomerName', 'KMeans_Cluster', 'DBSCAN_Cluster', 'Agglo_Cluster']], on='CustomerName')
```

```python
# Cluster profiles
numeric_columns = ['Recency', 'Frequency', 'Monetary']

kmeans_profiles = rfm.groupby('KMeans_Cluster')[numeric_columns].mean()
dbscan_profiles = rfm.groupby('DBSCAN_Cluster')[numeric_columns].mean()
agglo_profiles = rfm.groupby('Agglo_Cluster')[numeric_columns].mean()
```

```python
print("K-Means Cluster Profiles:\n", kmeans_profiles)
```

```python
print("DBSCAN Cluster Profiles:\n", dbscan_profiles)
```

```python
print("Agglomerative Clustering Profiles:\n", agglo_profiles)
```

```python
# Visualize the clusters for K-Means (as an example)
sns.pairplot(rfm, hue='KMeans_Cluster', vars=numeric_columns)
plt.show()
```

## Actionable Insights and Recommendations

```python
# Analyze clusters to provide actionable insights for K-Means (as an example)
for cluster in kmeans_profiles.index:
    print(f"K-Means Cluster {cluster} Profile:")
    print(kmeans_profiles.loc[cluster])
    print("Actionable Insights:")
    if kmeans_profiles.loc[cluster, 'Recency'] > rfm['Recency'].mean():
        print(" - These customers haven't purchased recently. Consider re-engagement campaigns.")
    if kmeans_profiles.loc[cluster, 'Frequency'] > rfm['Frequency'].mean():
        print(" - These customers purchase frequently. Consider loyalty programs or upsell opportunities.")
    if kmeans_profiles.loc[cluster, 'Monetary'] > rfm['Monetary'].mean():
        print(" - These customers spend a lot. Consider premium offerings or exclusive deals.")
    print("\n")
```
